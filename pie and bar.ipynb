```python
def pie_chart(*values, label=None, autopct:str='%1.1f%%', colors=None,explode=None, shadow=False, size="smaller", title='Pie Chart'):

    import matplotlib.pyplot as plt
    import numpy as np


    fig, ax = plt.subplots(figsize=(15, 8))
    ax.pie(x=np.ravel(values), labels=label, colors=colors,autopct=autopct, explode=explode, shadow=shadow, textprops={'size':size})
    ax.set_title(title)
    fig.set_size_inches(10, 5)

    plt.show()
```

```python
def pareto_chart(x, y):
    from matplotlib.axis import Axis
    import matplotlib.pyplot as plt
    import numpy as np
    cum_value = np.cumsum(y)
    fig, axe = plt.subplots(figsize=(10, 8))
    axe.bar(height=y, x=x)
    ax = axe.twinx()
    ax.plot(x, cum_value, color='red', marker="D", ms=6)
    Axis.get_minor_formatter(axe.yaxis)

    plt.xticks(rotation=45)  # Add this line to rotate the x-axis tick labels by 45 degrees.
    plt.show()

pareto_chart(religion,members)
```

</h1>

```python
def bar_plot(x, height, label:list , colors:list=[], width=0.8, bottom=None, *, rotation=45,align='center', legends:str="Chart Legend" ,data=None, **kwargs):
    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib.colors as mcolors
    import random
    all_colors = []
#    def random_color():
 #       color = np.random.randint(0, 6, size=3)
 #       return f'#{color[0]:02x}{color[1]:02x}{color[2]:02x}'
    def random_color():
        colors = list(mcolors.CSS4_COLORS.keys())
        return mcolors.hex2color(random.choice(colors))
    if len(colors) > 0:
        all_colors = [f'tab:{color}' for color in colors]
    else:
        for i in range(len(x)):
            all_colors.append(random_color())
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(x, height, label=label, color=all_colors, width=width, align= align, data=data )
    plt.xticks(rotation=rotation)
    if 'ylabel' in kwargs and 'title' and 'xlabel' in kwargs and 'legend' in kwargs:
        ax.set_ylabel(kwargs['ylabel'])
        ax.set_xlabel(kwargs['xlabel'])
        ax.set_title(kwargs['title'])
    ax.legend(title=legends)

    plt.show()
```

```python
def dot_plot(*data_distribution, **kwargs):
    import matplotlib.pyplot as plt
    from matplotlib.axis import Axis
    import numpy as np
    scatter_x = []
    scatter_y = []
    colors = []
    values, counts = np.unique(data_distribution, return_counts=True)
    for i, value in enumerate(values):
        for count in range(1, counts[i]+1):
            scatter_x.append(value)
            scatter_y.append(count)
            colors.append(value)

    # draw dot plot using scatter() 
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.scatter(scatter_x, scatter_y, c=colors)
    ax.get_yaxis().set_visible(False)
    # Optional - show all unique values on x-axis. 
    # Matplotlib might hide some of them  
    ax.set_xticks(values)
    # plt.gca().set_xticks(values)
    #set minor ticks label
    # plt.minorticks_on()
    #set major ticks label
    # plt.tick_params(which='major', length=10)
    plt.show()
                
            
```

```python
import matplotlib.pyplot as plt
import numpy as np

def line_chart(x1:list, y1:list, fmt1=None, x2:list=None, y2:list=None, fmt2=None, **kwargs ):
    x1_cor = np.array(x1)
    y1_cor= np.array(y1)
    label1 = kwargs['label1'] if 'label1' in kwargs else ''
    label2 = kwargs['label2'] if 'label2' in kwargs else ''
    xlabel = kwargs['xlabel'] if 'xlabel' in kwargs else ''
    ylabel = kwargs['ylabel'] if 'ylabel' in kwargs else ''
    title = kwargs['title'] if 'title' in kwargs else ''
    if x2 is None and y2 is None:
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.plot(x1_cor, y1_cor, label = label1 if fmt1==None else fmt1)
    else:
        x2_cor = np.array(x2)
        y2_cor= np.array(y2)
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.plot(x1_cor, y1_cor, label=label1 if fmt1==None else fmt1)
        ax.plot(x2_cor, y2_cor,label=label2 if fmt2==None else fmt2)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    if label1 != '' or label2 != '':
        ax.legend()
    plt.grid()
    plt.show()
```

```python
day = np.array([1, 2, 3, 4, 5], dtype=int)
time = np.array([45, 43, 46, 32, 25])
datas={
    'xlabel':'Days',
    'ylabel':'Time',
    'title':'Description'
      }
line_chart(x1=day, y1=time, **datas)
```
